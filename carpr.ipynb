```python
#import libraries


import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")


#load dataset

file_path = "C:/Users/ANSIL SHARAFUDEEN/Downloads/car data.csv"
df = pd.read_csv(file_path)
df
```

```python

#check missing values

print("\nMissing Values:\n",df.isnull().sum())
```

```python

#drop unecessary column

df.drop(['mileage(km/ltr/kg)','engine','max_power','seats'], axis=1, inplace=True)
```

```python
#data preprocessing

current_year = 2026
df['Car_Age'] = current_year - df['year']
```

```python
#data visualization
#graph 1-fuel type distribution


plt.figure()
sns.countplot(x='fuel', data=df)
plt.title("Fuel Type Distribution")

print("\nObservation 1:")
print("Diesel cars are most common, followed by petrol.")

plt.show()
```

```python
#graph 2-transmission type

plt.Figure()
sns.countplot(x='transmission', data=df)
plt.title("Transmission Distribution")

print("\nObservation 2:")
print("Manual Cars dominate the dataset.")

plt.show()
```

```python
#graph 3-selling price distribution

plt.Figure()
sns.histplot(df['selling_price'], bins=30)
plt.title("Selling Price Distribution")

print("\nObservation 3:")
print("Most cars are sold at lower price ranges.")

plt.show()
```

```python
#graph 4-car age vs selling price


plt.Figure()
sns.scatterplot(x='Car_Age', y='selling_price', data=df)
plt.title("Car Age Vs Selling Price")

print("\nObservation 4:")
print("Older cars tend to have lower selling prices.")

plt.show()
```

```python
#final conclusion

print("\n===================================")
print("FINAL CONCLUSION")
print("===================================")

print("\n1.Diesel cars are more common in the dataset.\n2.Manual transmission vehicles dominate.\n3.Selling price decreases as car age increases.\n4.Present price is a strong indicator of selling price.")
```
